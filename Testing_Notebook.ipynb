Let's try to generate a list of urls for extracting the house pricing for each Montreal borough

```python
# import pandas as pd # we gonna be working with dataframes
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
```

```python
url = "https://en.wikipedia.org/wiki/Urban_agglomeration_of_Montreal#Subdivisions"
data  = requests.get(url).text
mtl_soup = BeautifulSoup(data, 'html.parser')
print(mtl_soup.prettify())
```

```python
boroughs_soup=[]
boroughs_names=[]
boroughs_names_for_URLs=[]
boroughs_soup = mtl_soup.find_all('ul')[1]('a')
boroughs_soup.append((mtl_soup.find_all('ul')[3]('a'))[0])
for i in range(len(boroughs_soup)):
    # creates the list of boroughs original names 
    boroughs_names.append(boroughs_soup[i].text.replace('–', '-').replace('Mount Royal', 'Mont-Royal').replace('Montreal West', 'Montréal-Ouest').replace('Côte Saint-Luc', 'Côte-Saint-Luc') )
    # creates the list of boroughs for making URLs, dealing with french characters and special cases
    boroughs_names_for_URLs.append(boroughs_soup[i].text.replace('é', 'e').replace("'", '-').replace('ô', 'o').replace('Î', 'I').replace('â', 'a').replace('è', 'e').replace(' ', '-').replace('–', '-').replace('Verdun', 'Verdun-Ile-des-Soeurs').replace('Mount-Royal', 'Mont-Royal').replace('Montreal-West', 'Montreal-Ouest') )
boroughs_names.remove('Montreal')
boroughs_names_for_URLs.remove('Montreal')
boroughs_names_for_URLs
```

```python
boroughs_names
```

```python
# import urllib.request
median_housing_prices = []

for i, bor in zip(range(len(boroughs_names_for_URLs)), boroughs_names_for_URLs):
    cell = {}
    # print(bor.lower())
    if requests.get(f'https://www.centris.ca/en/tools/real-estate-statistics/montreal-island/{bor.lower()}', headers={'User-Agent': 'Mozilla/5.0'}).status_code == 200:
        webpage_as_text = requests.get(f'https://www.centris.ca/en/tools/real-estate-statistics/montreal-island/{bor.lower()}', headers={'User-Agent': 'Mozilla/5.0'}).text
    elif requests.get(f'https://www.centris.ca/en/tools/real-estate-statistics/montreal-island/montreal-{bor.lower()}', headers={'User-Agent': 'Mozilla/5.0'}).status_code == 200:
        webpage_as_text = requests.get(f'https://www.centris.ca/en/tools/real-estate-statistics/montreal-island/montreal-{bor.lower()}', headers={'User-Agent': 'Mozilla/5.0'}).text
    else:
        print(f"Can't find the correct URL for scraping. Please check the URL for: {bor.lower()}")
        print(f"First tried to use: https://www.centris.ca/en/tools/real-estate-statistics/montreal-island/{bor.lower()} but failed")
        print(f"Then tried to use: https://www.centris.ca/en/tools/real-estate-statistics/montreal-island/montreal-{bor.lower()} but also failed")
        continue
    webpage_as_soup = BeautifulSoup(webpage_as_text, 'html.parser')
    cell['Borough'] = boroughs_names[i]
    cell['singleFamily_2nd_Q_2021'] = webpage_as_soup.find(id='stat61').text
    cell['singleFamily_last_4Qs'] = webpage_as_soup.find(id='stat63').text
    cell['condo_2nd_Q_2021'] = webpage_as_soup.find(id='stat29').text
    cell['condo_last_4Qs'] = webpage_as_soup.find(id='stat31').text
    cell['plex_2nd_Q_2021'] = webpage_as_soup.find(id='stat45').text
    cell['plex_last_4Qs'] = webpage_as_soup.find(id='stat47').text
    median_housing_prices.append(cell)
```

```python
metro_stations_json = json.load(open('stm_arrets_sig.geojson', encoding='utf-8-sig'))
stations_list=[]
for station in metro_stations_json['features']:
    if station['properties']['loc_type'] == 0 and (station['properties']['route_id'] == '1' or station['properties']['route_id'] == '2' or station['properties']['route_id'] == '4' or station['properties']['route_id'] == '5'):
        station['properties'].pop('stop_id', None)
        station['properties'].pop('stop_code', None)
        station['properties'].pop('wheelchair', None)
        station['properties'].pop('loc_type', None)
        station['properties'].pop('service_id', None)
        if station['properties']['route_id'] == '1':
            station['properties']['route_id'] = 'Green line'
        elif station['properties']['route_id'] == '2':
            station['properties']['route_id'] = 'Orange line'
        elif station['properties']['route_id'] == '4':
            station['properties']['route_id'] = 'Yellow line'
        elif station['properties']['route_id'] == '5':
            station['properties']['route_id'] = 'Blue line'
        stations_list.append(station)
    # else:
    #     station.pop('type', None)
    #     station.pop('geometry', None)
    #     station.pop('properties', None)
metro_stations_json['features'] = stations_list


metro_stations_json
```

```python
def getNearbyVenues(boroughs, radius=500):
    
    venues_list=[]
    for bor_name in boroughs:
        #print(bor_name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&near={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            f"{bor_name}, QC, CA", 
            LIMIT)
        
        # make the GET request to FourSquare API
        code = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'}).json()['meta']['code']
        results = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'}).json()["response"]['groups'][0]['items']
        if code == 200:
            print(f" {len(results)} venues for {bor_name} found in FourSquare")
            # print(url)
            # print(results)
        else:
            print(f"{bor_name} NOT FOUND IN FOURSQUARE. CHECK FOR A BOROUGH NAME CORRECTION")
            # print(url)
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            bor_name, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['location']['formattedAddress'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Borough', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Address', 
                  'Venue Category']
    
    return(nearby_venues)
```
